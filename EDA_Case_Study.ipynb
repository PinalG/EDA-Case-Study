#### We cannot simply delete this as there can be exception in incomes so we will create separate groups to distribute data imbalance

```python
def new_category(r):
    if r>=600000:
        return 'high'
    if r>=200000:
        return 'medium'
    return 'low'

appl['INCOME_CATEGORY'] = appl.AMT_INCOME_TOTAL.apply(lambda x: new_category(x))
appl.head()
```

#### we can do same for credit amount for imbalanced data

```python
plt.figure()
plt.boxplot(appl.AMT_CREDIT)
plt.show()
```

```python
appl.AMT_CREDIT.describe()
```

```python
appl['CREDIT_CATEGORY'] = appl.AMT_CREDIT.apply(lambda x: new_category(x))
appl.head()
```
